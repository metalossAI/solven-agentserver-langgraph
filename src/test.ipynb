```python
# Self-contained test: import, define and run the async main safely in notebooks (with guarded output)
import asyncio
from mcps import catastro_mcp

async def main():
    # Replace 'catastro' with your server_name if different
    try:
        tools = await catastro_mcp.get_tools(server_name="catastro")
        # Print concise summary only to avoid huge outputs
        count = len(tools) if hasattr(tools, '__len__') else 'unknown'
        print(f'got tools: count={count}, type={type(tools)}')
        if hasattr(tools, '__len__') and len(tools) > 0:
            first = tools[0]
            preview = str(first)[:400]
            print('first tool preview:', preview)
        return tools
    except Exception as e:
        # Print only exception type and short message to avoid massive dumps
        import traceback
        msg = repr(e)
        print('Exception calling get_tools:', type(e).__name__, msg[:400])
        return None

# Execute using top-level await when possible
try:
    loop = asyncio.get_running_loop()
    loop_running = loop.is_running()
except RuntimeError:
    loop_running = False

if loop_running:
    result = await main()
else:
    result = asyncio.run(main())
print('result_type:', type(result))
```
